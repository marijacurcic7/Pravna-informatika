```python
import os
import re
import nltk
import pandas as pd
```

```python
rootdir = 'data'
filepaths = []

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        filename = os.path.join(subdir, file)
        filepaths.append(filename)
```

```python
data = []
data_sentences = []
for file in filepaths:
    f = open(file, "r", encoding="utf8")
    sadrzaj = f.read()
    sadrzaj = sadrzaj.lower()
    sadrzaj = sadrzaj.replace("\n", " ")
    sadrzaj = sadrzaj.replace("*", "")
    sadrzaj = " ".join(sadrzaj.split()) #replace multiple whitespace characters with single space
    tokens = nltk.sent_tokenize(sadrzaj)
    row = [file, tokens]
    data.append(row)
    for t in tokens:
        row_sent = [t, '', file]
        data_sentences.append(row_sent)
```

```python
df = pd.DataFrame(data=data, columns=['filename', 'sentences'])
df #red je fajl sa listom recenica
```

```python
df_sentences = pd.DataFrame(data=data_sentences, columns=['sentence', 'label', 'filename'])
df_sentences #red je jedna recenica sa labelom i nazivom fajla u kojem se nalazi
#sve labele su za sad prazna, bice 1 ili 0
```

```python
data = []
data_sentences = []
for file in filepaths:
    f = open(file, "r", encoding="utf8")
    sadrzaj = f.read()
    sadrzaj = sadrzaj.lower()
    sadrzaj = sadrzaj.replace("\n", " ")
    sadrzaj = sadrzaj.replace("*", "")
    sadrzaj = sadrzaj.replace("br.", "br")
    sadrzaj = sadrzaj.replace("čl.", "čl")
    sadrzaj = sadrzaj.replace("tač.", "tač")
    sadrzaj = sadrzaj.replace("kt.", "kt")
    sadrzaj = " ".join(sadrzaj.split()) #replace multiple whitespace characters with single space
    tokens = nltk.sent_tokenize(sadrzaj)
    row = [file, tokens]
    data.append(row)
    for t in tokens:
        row_sent = [t, '', file]
        data_sentences.append(row_sent)
```

```python
df_sentences1 = pd.DataFrame(data=data_sentences, columns=['sentence', 'label', 'filename'])
df_sentences1
```
