```python
list_sent_tkn = []
for sent in list_sentences:
    list_sent_tkn.append(word_tokenize(sent))
w2v_model = Word2Vec(list_sent_tkn, min_count=1, vector_size= 50, workers=3, window=3)
```

```python
words = list(w2v_model.wv.index_to_key)
print("Words: ", len(words))
```

```python
word_vec_dict={}
for word in words:
    word_vec_dict[word]=w2v_model.wv.get_vector(word)
print("The no of key-value pairs : ",len(word_vec_dict)) # should come equal to vocab size
```

```python
maxi=-1
for i,rev in enumerate(list_sentences):
    tokens = rev.split()
    if(len(tokens)>maxi):
        maxi=len(tokens)
```

```python
tok = Tokenizer(filters='')
tok.fit_on_texts(list_sentences)
vocab_size1 = len(tok.word_index) + 1
encd_rev_labeled = tok.texts_to_sequences(labeled_df['sentence'])
encd_rev_unlabeled = tok.texts_to_sequences(unlabeled_df['sentence'])
```

```python
print(vocab_size1)
```

```python
pad_rev_labeled = pad_sequences(encd_rev_labeled, maxlen=MAX_SEQUENCE_LENGTH, padding='post')
print(pad_rev_labeled)
print(pad_rev_labeled.shape)
```

```python
pad_rev_unlabeled = pad_sequences(encd_rev_unlabeled, maxlen=MAX_SEQUENCE_LENGTH, padding='post')
print(pad_rev_unlabeled)
print(pad_rev_unlabeled.shape)
```

```python
embed_matrix = np.zeros(shape=(vocab_size1, embedding_size))
for word,i in tok.word_index.items():
    embed_vector = word_vec_dict.get(word)
    if embed_vector is not None:
        embed_matrix[i] = embed_vector
```

## BiLSTM obucavanje

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model
```

```python
train_size=0.8

y_labeled = np.array(labeled_df['label'].tolist())

leinst = LabelEncoder()
yinst = leinst.fit_transform(y_labeled)

ycat_labeled = to_categorical(labeled_df['label'])  # one hot target as required by NN
X_train, X_validation, y_train, y_validation = train_test_split(pad_rev_labeled, ycat_labeled, test_size=0.1, shuffle=True, random_state=42)

X_unlabeled = pad_rev_unlabeled

num_classes = len(leinst.classes_)

print("X_train:", X_train.shape, " y_train:",y_train.shape)
print("X_validation:", X_validation.shape, " y_validation:", y_validation.shape)
print("X_unlabeled:", X_unlabeled.shape)

print("num_classes:", num_classes)
print("Classes:", leinst.classes_)
```

```python
model = Sequential()
# Keras Embedding layer with Word2Vec weights initialization
model.add(Embedding(input_dim=vocab_size1,
                    output_dim=embedding_size,
                    weights=[embed_matrix],
                    input_length=maxi+1,
                    ))

model.add(Bidirectional(LSTM(30)))
# model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))
# model.add(Flatten())
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
stopper = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
```

```python
n0 = 0
n1 = 100
for i in range(0, 5):
    
    checkpointer = ModelCheckpoint(
                filepath="models/model_bilstm_" + str(i + 1) + "_best.hdf5",
                verbose=1, save_best_only=True)
    
    model.fit(X_train, y_train, epochs=50, batch_size=32,
                        validation_data=(X_validation, y_validation), 
                        callbacks=[checkpointer, stopper],
                        verbose=1)
    
    model = load_model("models/model_bilstm_" + str(i + 1) + "_best.hdf5")
    loss, acc = model.evaluate(X_validation, y_validation, verbose=1)
    print("Validation Accuracy: ", acc)
    print("Validation Loss: ", loss)
    
    X_train_old = X_train
    y_train_old = y_train
    X_validation_old = X_validation
    y_validation_old = y_validation
    
    X_test = X_unlabeled[n0:n1]
    
    n0 = n1
    n1 = n1 + 100
    
    predictions = model.predict([X_test])
    
    y_test = []
    for p in predictions:
        i = np.argmax(p)
        y_test.append(i)
        
    X_labeled = np.concatenate((X_train_old, X_validation_old, X_test), axis=0)
    
    ycat_test = to_categorical(y_test)
    y_labeled = np.concatenate((y_train_old, y_validation_old, ycat_test), axis=0)
    
    X_train, X_validation, y_train, y_validation = train_test_split(X_labeled, y_labeled, test_size=0.1, shuffle=True, random_state=42)
```

```python
loss, acc = model.evaluate(X_validation, y_validation, verbose=1)
print("Validation Accuracy: ", acc)
print("Validation Loss: ", loss)
```

## Klasterizacija

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#ucitavanje csv fajla koji je nastao pokretanjem Spajanje recenica sa argumentima.ipynb
#red je deo recenice koja je labelirana da sadrzi argument
df1 = pd.read_csv('split_by_zarez_df_cluster.csv')
```

```python
#odabir svih redova koji se odnose na jedan fajl
#klasteruju se delovi recenica na nivou jednog fajla
df_klasterovanje = df1[df1.filename == 'data\K - PRVOSTEPENI KRIVIČNI PREDMETI\presuda_text_10160.txt']
```

```python
vec = TfidfVectorizer(stop_words=None)
vec.fit(df_klasterovanje.sentence.values)
features = vec.transform(df_klasterovanje.sentence.values)
```

```python
cls = MiniBatchKMeans(n_clusters=3, random_state=None)
cls.fit(features)
```

```python
cls.predict(features)
cls.labels_
```

```python
len(cls.labels_)
```

```python
#cuvanje prediktovanih klastera u zasebnoj koloni
df_klasterovanje['cluster'] = pd.Series(cls.labels_, index=df_klasterovanje.index)
```
